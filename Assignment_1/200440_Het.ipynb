```julia
import Pkg; 
Pkg.add("Distributions")
Pkg.add("PyPlot")
using Distributions
lambda = 10
iter = 1e4
X = []
for j in 1:iter
    U = rand(Uniform(0,1))
    i = 0
    p = exp(-lambda)
    A = p
    while U>=A
        p = (lambda/(i+1))*p
        A = A + p
        i = i + 1
    end
    push!(X,i)
end
X
Pkg.add("Plots")
using PyPlot
using Plots
display(histogram(X))
mean(X)
```
