```julia
import Pkg; Pkg.add("Distributions")
import Pkg; Pkg.add("Plots")
using Plots
using Distributions
repeat = 1e4

arr = []
for k in 1:repeat
    i = 0
    p = exp(-7)
    A = p
    U = rand(Uniform(0,1))
    while U>=A
        p = (7/(i+1))*p
        A = A + p
        i = i + 1
    end
    push!(arr,i)
end
display(histogram(arr))
```

```julia
import Pkg; Pkg.add("Distributions")
import Pkg; Pkg.add("Plots")
using Plots
using Distributions
n = 40
p = 0.25
max = 0
accept = 0
repeat = 1e4
arr = []
for i in 1:n
    c = (binomial(n,i) * ((1-p)^(n-i))*(p^i)/((exp(-1*2)*(2^i))/factorial(big(i))))
    if max<c
        max = c
    end
end
for k in 1:repeat
        accept = 0
        while accept == 0
                U = rand(Uniform(0,1))
                Y = rand((0:n))
                ratio = ((1/max)*(binomial(n,Y) * ((1-p)^(n-Y))*(p^Y))/((exp(-1*2)*(2^Y))/factorial(big(Y))))
                if U < ratio
                    accept = 1
                    push!(arr,Y)
                end
        end
end
display(histogram(arr)) 
```
