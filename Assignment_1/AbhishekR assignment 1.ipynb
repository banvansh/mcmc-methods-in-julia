```julia
import Pkg; Pkg.add("Distributions")
import Pkg;
Pkg.add("Plots")
function poisson(lambda)
    x=0
    p=exp(-lambda)
    s=p    
    u=rand(Float64)  
      
    while x>-1
        if u<s
            return x
        else
            x=x+1    
            p=p*lambda/x    
            s=s+p
        end
    end
end
arr=[]

for i in 0:500
    push!(arr,poisson(10))
end

using Plots
histogram(arr)
```
