Pr(X=1) = Pr(C1=1 && C2=1|C1=1)/(Pr(C1=1 && C2=1|C1=1)+Pr(C1=0 && C2=1|C1=0))
        = (cypy/(cy+cx))/(cypy/(cy+cx) + cxpx/(cy+cx))
        = cypy/(cxpx+cypy)
Clearly, This is the pmf of Bern(cypy/(cxpx+cypy)). 


```julia
using StatsBase
using Distributions
using Plots
using SpecialFunctions
```

```julia
function Bernoulli_inv_trans(p)
    U = rand(Uniform(0, 1))
    if U < p
        X = 0
        return X
        break
        
    else
        X = 1
        return X
        break
    end
end
cx = 2
cy = 3
px = 1/2
py = 1/3
i = 0
function Required_inv_transform(cx, cy, px, py)
    arr = Array{Int64}(undef, 2)
    while 1<2
        i = i+1
        C1 = Bernoulli_inv_transform(cx/(cx+cy))
        if C1 = 1
            C2 = Bernoulli_inv_trans(py)
            if C2 = 1
                arr[1]= 1
                arr[2]= i
                return arr
                break
            else
                continue
                break
            end
        else
            C2 = Bernoulli_inv_trans(px)
            if C2 = 1
                arr[1]= 0
                arr[2]= i
                return arr
                break
            else
                continue
                break
            end
        end
    end
end
```

```julia
function sphere_ar()
    accept = 0 
    counter = 0
    while accept == 0
        counter = counter + 1
        prop = []
        append!(prop, rand(Uniform(-1,1)))
        append!(prop, rand(Uniform(-1,1)))
        posn = prop[1]^2+prop[2]^2 < 1
        if posn
            accept = 1
            return prop, counter
        else
            accept = 0
        end
    end
end
Max=zeros(0,2)
count=0
n=0
while count<10000
    x,num=sphere_ar()
    n=n+num
    Max=vcat(Max, reshape(x,(1,2)))
    count=count+1
end

println("Avg Iteration:",n/10000)
g=1
f=g/pi
f=f*(4)
println("Expected Iteration:",f)


plot(Max[:,2],Max[:,1],seriestype = :scatter,)
```
