<h2><center>Q2 Sample uniformly from a p-dimensional sphere</center></h2>

```julia
function generatesphere(p)
    loop_complete = 1
    temp_coordinates = []
    final_coordinates = []
    tot = 1
    for i in 1:p
        a = -1 + 2 * rand()
        tot = tot - a * a
        if(tot < 0)
            loop_complete = 0
            break
        else
            push!(temp_coordinates, a)
        end
    end
    if(loop_complete == 1)
        final_coordinates = temp_coordinates
    end
    return final_coordinates
end
```

```julia
# generating uniform sample from circle
X = []
Y = []
for j in 1:10000
    arr = generatesphere(2)
    if(isempty(arr) == false)
        push!(X, arr[1])
        push!(Y, arr[2])
    end
end
```
