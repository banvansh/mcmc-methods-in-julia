$\sigma_g^2= Var_g(\frac{h(Z)\pi(Z)}{g(Z)})$

```julia
function find_mean_and_variance(x)
    est_mean=0
    arr=[]
    for i in 1:x
        push!(arr,(norm_samp[i]*pdf(Cauchy(0.0,1.0),norm_samp[i])/pdf(Normal(0.0,1.0),norm_samp[i])))
        est_mean+=arr[i]
    end
    println(est_mean/=n)
    println(var(arr))
    return var(arr)
end
```

```julia
find_mean_and_variance(10000)
```

```julia
var_arr = []
for i in 1:10000
    push!(var_arr, find_mean_and_variance(i))
end
```

$Var_g[{\hat{\theta}}] = \sigma_g^2\left( 1 + \frac{Var_g(w(Z))}{E_g(w(Z))^2}\right)$

```julia
#variance functio
function find_variance(x)
    n1 = x
    arr_norm = rand(Normal(1, 0), n1)
    prod_arr = []
    for i in 1:n1
        prod = 1
        for j in 1:50
            prod*=((1 + ((arr_t[j] - arr_norm[i])*(arr_t[j] - arr_norm[i]))/v)^((v+1)/2))
        end
        push!(prod_arr, prod)
    end
    est_mean_1 = 0
    for i in n1
       est_mean_1+=(prod_arr[i]*arr_norm[i]) 
    end
    est_mean_1/=(sum(prod_arr))
    new_arr_1 = []
    for i in 1:n1
       push!(new_arr_1, prod_arr[i]*arr_norm[i]) 
    end
#     variance = var(new_arr_1)*(1 + var(prod_arr)/(mean(prod_arr)*mean(prod_arr)))
    variance = var(new_arr_1)*(1 + var(prod_arr)/(mean(prod_arr))*(mean(prod_arr)))
    return(variance)
end
```

```julia
var_arr_2 = []
for i in 1:40
    push!(var_arr_2, find_variance(i*1000))
end
plot(var_arr_2)
```
